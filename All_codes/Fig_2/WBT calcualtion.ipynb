```python
import os
import gc
import xarray as xr
import numpy as np
import metpy.calc as mpcalc
from metpy.units import units
from tqdm import tqdm

directory_path = '/content/drive/MyDrive/AP_HW/ERA5 Data/t2m_d2m_sp'

for filename in tqdm(sorted(os.listdir(directory_path))):
    if filename.endswith('.nc'):
        file_path = os.path.join(directory_path, filename)
        # print(filename)
        year = filename[-7:-3]
        if int(year)>=1971 and int(year)<=1980:
          # Open dataset
          ds = xr.open_dataset(file_path)
          start_time=str(np.datetime64('{}-03-01T00:00:00.000000000'.format(year)) - np.timedelta64(6, 'h'))
          end_time=str(np.datetime64('{}-11-30T17:00:00.000000000'.format(year)))
          time_range=slice(start_time,end_time)
          lat_range=slice(26.5,20.75)
          lon_range=slice(88,92.5)
          ds = ds.sel(latitude=lat_range, longitude=lon_range, time=time_range)
          ds['time']=ds['time']+np.timedelta64(6,'h')
          ds
          # Calculate wet bulb temperature
          wbt = mpcalc.wet_bulb_temperature(ds.sp * units.Pa, ds.t2m * units.kelvin, ds.d2m * units.kelvin)
          wbt.name = 'wbt'
          wbt.attrs['units'] = 'kelvin'
          wbt = wbt.to_dataset()

          # Save the output to NetCDF
          wbt.to_netcdf(f'/content/drive/MyDrive/AP_HW/Scripts-ll/WBT/wbt_files/wbt-{year}.nc')
          print(f"Finished caculating wbt-{year}.nc")
```
