# Percenntage change Calculation

```python
def percentage_change(value1,value2):
  difference=value2-value1
  percentage=(difference/value1)*100
  return percentage
```

```python
def mask_bd(ds):
  gdf = gpd.read_file("/content/drive/MyDrive/Other_projects/random/forecast/gadm41_BGD_shp/gadm41_BGD_0.shp")
  gdf = gdf.to_crs("EPSG:4326")
  ds = ds.rio.write_crs("EPSG:4326", inplace=True)
  masked_ds = ds.rio.clip(gdf.geometry.apply(mapping), gdf.crs, drop=True)
  return masked_ds


def spatial_mean(ds):
  weights=np.cos(np.deg2rad(ds.latitude))
  weights.name='weights'
  ds_weighted = ds.weighted(weights)
  BD_mean = ds_weighted.mean(['longitude','latitude'])
  return BD_mean
```

```python
# --------------------------------------------------------------------
# Pre-monsoon (MAM) total cloud cover change during events
# --------------------------------------------------------------------

# Open climatology dataset for MAM (pre-monsoon) TCC
clim = xr.open_dataset('/content/drive/MyDrive/AP_HW/Scripts-ll/climatology/data/tcc/tcc_climatology-mam.nc')
clim

# Open TCC event dataset for MAM and take mean over all events
ds_events = xr.open_dataset('/content/drive/MyDrive/AP_HW/Scripts-ll/tcc/data/tcc_events-mam.nc')
ds_events = ds_events.mean(dim='event_no', skipna=True)
ds_events

# Apply Bangladesh mask to both climatology and event datasets
masked_events = mask_bd(ds_events)
masked_clim = mask_bd(clim)

# Compute spatial average over Bangladesh for events and climatology
BD_mean_events = spatial_mean(masked_events)
BD_mean_clim = spatial_mean(masked_clim)

# Calculate percentage change in TCC during MAM heatwave events compared to climatology
tcc_mam_percentage = percentage_change(BD_mean_clim.tcc.values, BD_mean_events.tcc.values)
tcc_mam_percentage
print('change_in_tcc_premonsoon:', tcc_mam_percentage, '%')

# --------------------------------------------------------------------
# Monsoon (JJAS) total cloud cover change during events
# --------------------------------------------------------------------

# Open climatology dataset for JJAS (monsoon) TCC
clim = xr.open_dataset('/content/drive/MyDrive/AP_HW/Scripts-ll/climatology/data/tcc/tcc_climatology-jjas.nc')
clim

# Open TCC event dataset for JJAS and take mean over all events
ds_events = xr.open_dataset('/content/drive/MyDrive/AP_HW/Scripts-ll/tcc/data/tcc_events-jjas.nc')
ds_events = ds_events.mean(dim='event_no', skipna=True)
ds_events

# Apply Bangladesh mask to both climatology and event datasets
masked_events = mask_bd(ds_events)
masked_clim = mask_bd(clim)

# Compute spatial average over Bangladesh for events and climatology
BD_mean_events = spatial_mean(masked_events)
BD_mean_clim = spatial_mean(masked_clim)

# Calculate percentage change in TCC during JJAS heatwave events compared to climatology
tcc_jjas_percentage = percentage_change(BD_mean_clim.tcc.values, BD_mean_events.tcc.values)
tcc_jjas_percentage
print('change_in_tcc_monsoon:', tcc_jjas_percentage, '%')
```
