# Ept calculation and saving

```python
import os
from tqdm.auto import tqdm
directory_path='/content/drive/MyDrive/AP_HW/ERA5 Data/t2m_d2m_sp'

for filename in tqdm(sorted(os.listdir(directory_path))):
  if filename.endswith('.nc'):
    filepath=os.path.join(directory_path,filename)
    ds=xr.open_dataset(filepath)
    year=int(filename[11:-3])
    lat_range=slice(26.5,20.75)
    lon_range=slice(88,92.5)
    ds=ds.sel(latitude=lat_range,longitude=lon_range)
    theta_e = mpcalc.equivalent_potential_temperature(ds.sp*units.Pa,ds.t2m*units.kelvin,ds.d2m*units.kelvin)
    theta_e.name='ept'
    theta_e.attrs['units']= 'K'
    theta_e.attrs['long_name']= 'Equivalent potential temperature'
    theta_e.to_netcdf(f'/content/drive/MyDrive/AP_HW/ERA5 Data/EPT/ept-{year}.nc')

```
